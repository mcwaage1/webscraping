```python
def get_words():
    list_of_urls =  ['a',
                     'b',
                     'c',
                     'd',
                     'e',
                     'f',
                     'g',
                     'h',
                     'i',
                     'j',
                     'k',
                     'l',
                     'm',
                     'n',
                     'o',
                     'p',
                     'q',
                     'r',
                     's',
                     't',
                     'u',
                     'v',
                     'w',
                     'x',
                     'y',
                     'z',
                     'special-character-number']
    
    for char in list_of_urls:
        globals()[f'{char}_df'.replace('-', '_')] = pd.DataFrame()
        
    driver = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver')

    list_of_words = []
    for char in list_of_urls[:-1]:
        print(f"Getting words for character: {char}")
        url = 'https://dictionary.apa.org/browse/' + char
        driver.delete_all_cookies()
        driver.get(url)
        c = driver.page_source
        soup = BeautifulSoup(c, "html.parser")
        for word in soup.find_all('a'):
            list_of_words.append(word.get('href'))
        [list_of_words.remove(i) for i in list_of_words if i == None]
        globals()[f'{char}_df'][f'{char}'] = pd.Series([i for i in list_of_words if i[0:4] != 'http' if i[0] == f'{char}'])
        print(f"Finished getting words for character: {char}")
        print(globals()[f'{char}_df'][f'{char}'].head())
        print()
        
    list_of_words = []
    for char in list_of_urls[-1:]:
        print(f"Getting words for character: {char}")
        url = 'https://dictionary.apa.org/browse/' + char
        driver.delete_all_cookies()
        driver.get(url)
        c = driver.page_source
        soup = BeautifulSoup(c, "html.parser")
        for word in soup.find_all('a'):
            list_of_words.append(word.get('href'))
        [list_of_words.remove(i) for i in list_of_words if i == None ]
        globals()[f'{char}_df'][f'{char}'.replace('-', '_')] = pd.Series([i for i in list_of_words if i[0] != '/' if i[0:4] != 'http'])
        print(f"Finished getting words for character: {char}")
        print(globals()[f'{char}_df'][f'{char}'.replace('-', '_')].head())
        print()

    return df
```

```python
def get_definitions(df=df):
    driver = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver')
    
    definitions = []
    for col in df.columns:     
        print(f"Creating '{col}_definitions_dict'") 
        globals()[f'{col}_definitions_dict'] = {}
        
        dictionary_list = []
        for index, word in df[col][:2].items():
            if pd.notna([word]):
                print(f"Getting definition for: {word}")
                url = f'https://dictionary.apa.org/{word}'
                driver.delete_all_cookies()
                driver.get(url)
                WebDriverWait(driver, timeout=4).until(lambda d: d.find_element_by_tag_name("dd"))

                c = driver.page_source
                soup = BeautifulSoup(c, "html.parser")
                soup = soup.find('app-definition')

                definition = []
                if soup.find('hw'):
                    for text in soup.find('dd'):
                        definition.append(text.get_text())
                        definition = [''.join(definition)]
                        definition = [i.replace("\n            ", " ").replace("\n", " ") for i in definition]
                        
                    for term in soup.find('hw'):
                        term = term.get_text()

                    dictionary = {"term": term,
                                  "definition": definition[0],}
                    dict_list = {word: dictionary}
                    dictionary_list.append(dict_list)
                    print(f"Added {word} and definition to '{col}_definitions_dict' \n")
                
                
                elif soup.find('em'):
                    definition = []
                    terms = []
                    if soup.find_all('dd'):
                        for text in soup.find_all('dd'):
                            definition.append(text.get_text())
                            definition = [''.join(definition)]
                            definition = [i.replace("\n            ", " ").replace("\n", " ") for i in definition]
#                             print(definition)
                            
                    elif soup.find_all('p'):
                        for text in soup.find_all('p'):
                            definition.append(text.get_text())
                            definition = [''.join(definition)]
                            definition = [i.replace("\n            ", " ").replace("\n", " ") for i in definition]
#                             print(definition)

                    for term in soup.find_all('em'):
                        terms.append(term.get_text())
                        term = [', '.join(terms)]
                        term = [i.replace("\n", " ") for i in term]
#                         print(term)
                                            
                    dictionary = {"term": term[0],
                                  "definition": definition[0],}
                    dict_list = {word: dictionary}
                    dictionary_list.append(dict_list)
                    print(f"Added {word} and definition to '{col}_definitions_dict' \n")

        globals()[f'{col}_definitions_dict'] = dictionary_list
        definitions.append(dictionary_list)
    
    globals()['definitions'] = {'dictionary' : definitions}
    print("Created all <col>_definitions_dict's")
    print("Finished scraping")
```
